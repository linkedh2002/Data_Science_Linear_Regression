#Uploading the Model File

```python
import pickle
with open('/content/Trip_Duration_Prediction_Model','rb') as f:
  reg=pickle.load(f)
```

#Getting the Input File from the Client and Predicting the output into a File

```python
import pandas as pd
df=pd.read_csv('/content/Trip_Duration_Prediction_Input.csv')
df
```

```python
time_df=reg.predict(df)
df['time(HRS)']=time_df
df
```

```python
import numpy as np
rounded_time=df['time(HRS)'].apply(np.floor)
```

```python
df['time(HRS)']=rounded_time
df
```

```python
df.to_csv('Trip_Duration_Prediction_Output.csv',index=False)
```
